```python
import pandas as pd

df = pd.read_csv('data/complete.csv')
dataset=df.iloc[:,:-91]
#dataset['position']=
to_drop = ['international_reputation', 'potential','stamina','strength','skill_moves', 'weak_foot', 'work_rate_att', 'work_rate_def', 'preferred_foot','gk_diving', 'gk_handling', 'gk_kicking', 'gk_positioning', 'gk_reflexes', 'ID','birth_date','special','photo','eur_value','eur_wage','eur_release_clause', 'name','club_logo','body_type', 'real_face','flag','nationality']
for i in to_drop:
    dataset=dataset.drop(i,axis=1)

dataset = dataset[df['league'].isin(['Spanish Primera División', 'French Ligue 1', 'German Bundesliga','English Premier League', 'Italian Serie A', 'Turkish Süper Lig','Portuguese Primeira Liga','German 2. Bundesliga'])]
#dataset = dataset.where(dataset['league'] in ['Spanish Primera División', 'French Ligue 1', 'German Bundesliga','English Premier League', 'Italian Serie A', 'Turkish Süper Lig','Portuguese Primeira Liga','German 2. Bundesliga'], axis=1)
leagues = pd.unique(dataset['league'])
dataset = dataset.dropna(subset=['league'])
dataset = dataset.reset_index()
#Tut nicht das, was ich möchte, aber ist trotzdem interessant:
dataset['BestPos'] = dataset[['rs', 'rw', 'rf', 'ram', 'rcm', 'rm', 'rdm', 'rcb', 'rb', 'rwb', 'st', 'lw', 'cf', 'cam', 'cm', 'lm', 'cdm', 'cb', 'lb', 'lwb', 'ls', 'lf', 'lam', 'lcm', 'ldm', 'lcb', 'gk']].idxmax(axis=1)

print(list(dataset.columns.values))
data
```

```python
position_mapping = {'rs':'Sturm', 'rw':'Sturm', 'rf':'Sturm', 'ram':'Mittelfeld', 'rcm':'Mittelfeld', 'rm':'Mittelfeld', 'rdm':'Mittelfeld', 'rcb':'Verteidigung', 'rb':'Verteidigung', 'rwb':'Verteidigung', 'st':'Sturm', 'lw':'Sturm', 'cf':'Sturm', 'cam':'Mittelfeld', 'cm':'Mittelfeld', 'lm':'Mittelfeld', 'cdm':'Mittelfeld', 'cb':'Verteidigung', 'lb':'Verteidigung', 'lwb':'Verteidigung', 'ls':'Sturm', 'lf':'Sturm', 'lam':'Mittelfeld', 'lcm':'Mittelfeld', 'ldm':'Mittelfeld', 'lcb':'Verteidigung', 'gk':'Torwart'}
position_mapping


def extract_position(x):
    for i in position_mapping:
        if x == i:
            return position_mapping[i]


dataset['BestPos'] = dataset['BestPos'].apply(extract_position)

dataset

player_info= dataset.iloc[:,:5]

X = dataset.iloc[:,5:-29]

y=dataset.iloc[:,-1]
```
